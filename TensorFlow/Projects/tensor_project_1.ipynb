```python
import tensorflow as tf
from tensorflow import keras
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train_flattened = X_train.reshape(len(X_train), 784)/255
X_test_flattened = X_test.reshape(len(X_test), 784)/255
```

```python
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,),activation = 'sigmoid')
])
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(X_train_flattened, y_train,epochs = 5)
```

```python
plt.matshow(X_test[0])
```

```python
y= model.predict(X_test_flattened)
```

```python
y[0]
```

```python
np.argmax(y[0])
```

```python
plt.matshow(X_test[1])
```

```python
np.argmax(y[1])
```

```python
y_predicted_label = [np.argmax(i) for i in y]
```

```python
cm = tf.math.confusion_matrix(y_test, y_predicted_label)
```

```python
cm
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(cm, annot=True,fmt = 'd')
plt.xlabel('Predicted label')
plt.ylabel('truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,),activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(X_train_flattened, y_train,epochs = 5)
```

```python
model.evaluate(X_test_flattened,y_test)
```
